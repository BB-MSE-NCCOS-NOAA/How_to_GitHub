#### What Not to Post on GitHub

When considering what content to post on GitHub, no distinction should be made between public and private GitHub repositories, regardless of ownership. GitHub private repositories are only meant to provide a closed environment for working projects not yet ready for public use and should never contain sensitive code. All NOAA projects hosted on GitHub should be suitable for

#### Transferring a Repository

Let's move on and assume that the repository in question is (1) does not contain any sensitve information (2), it's in working order and (3), it's ready to be transferred from a work GitHub account to the Organization account. Below is the easiest way of doing so:

From your account on GitHub, go to the repository being transferred, open the settings tab, scroll down to `Danger Zone`, and then select `Transfer`; from here you'll be able to type in the Organization account's name, verify, and then select `transfer`. 



You can perform a sanity check by going back to your GitHub account and checking to see if the repository is there (it won't be), then go to the Organization account and check if it's there (it will be). 

Easy!

Okay, so some time goes on and you realize that you need to modify the code, or there's reason to continue working on it. What now? Well so what we can do is `fork` the repository you just transferred to the Organization account. If you're not familiar `forking` is creating an identical copy of another repository and saving it to your GitHub account. This `forked` version is yours to work on, and it does not affect the original repository in any way. However, if you want those edits to be added to the original repository later on, you can merge them through a `pull request` (assuming the repository owner allows this). 

Note that `forking` is different from `cloning`: both create exact copies of the original repository at that specific point in time, but the edits in the `cloned` repository cannot be `merged` with the original with a `pull request`. So if you plan to contribute to the original repository, `fork` it; if you want a completely isolated version that's all for yourself, consider `cloning` it.

Below we can see the Organization account and the repository that we created, but transferred ownership. 



I want to continue working on this repo, so I'm going to `fork` it. This is as simple as clicking the `fork` button in the top-right corner within the repository itself. By doing this, a `forked` copy will be saved to my GitHub account. We can validate this by looking at the repository and seeing that it is indeed a `forked` copy.



Now I'm going to go to my GitHub Desktop application and `clone` this `forked` copy; this creates a version on my local computer that I can continue to work on.



Maybe I make a few edits and decide that this is better, working version of the repository that is currently on the Organization account. What we're going to do first is commit our edits in GitHub Desktop, and push to origin. This will just update the version of the repository that is currently on our GitHub account. 

The next step is to submit a `pull request`, which if accepted, will add those same edits to the version of the repository that is currently in the Organization account. 

To submit a `pull request`, we're going to go to our version of the repository that is currently on our GitHub account, and select `pull request`. Then we're going to select `New pull request`.



On this page GitHub will automatically create a `diff` showing the differences between our `forked` repository, and the base, which in this case is the original repository. This is a good time to look over all of the edits made to make sure that nothing will break as a result of the `merge`.




If everything looks good, we're gonna go ahead and select `Create pull request`. This will alter the owner of the original repository (which, might also be you) that there is a new `pull request` that needs their attention. If we switch perspectives and continue this tutorial from the Organizational account, we'll see that `pull request` in the repository under the button `pull requests`. 

As the Organization account it's important to look over these edits with scrutiny to ensure that you're only accepting edits that will add to the success of the repository. If you see somethings that need work, you can always add comments during the revision process telling the author of the `pull request` that you need them to fix XYZ before you're willing to accept and `merge`; this process is a lot like co-authoring a paper.



If you believe everything looks good, you can accept the `pull request` and then `merge`, which overwrites the original repository with all the edits from the other. Don't worry, if something breaks you can always roll back to a previous commit! Now that all version are up to date, users can `fork` this version, add edits, and submit addional `pull requests` to the version in the Organization account.
