### Welcome!

In this notebook we're going to create our first repository (or, "repo). You'll see that it's actually very easy and can be done in multiple ways.

- Create a repository on GitHub, pull to your local machine using GitHub Desktop
- Create a repository using GitHub Desktop, push to the main branch on GitHub
- Turn an existing folder containing code on your local machine into a remote repository, push to Github using GitHub Desktop

In this tutorial we're going to start with the easiest method (the first), and from experience you'll quickly learn the other two on your own.

#### Committing and Pushing a Repo(sitory)

Now if you look closely on the left-hand side, the Notebooks I've added are in red, which in this IDE means they contain edits that are uncommitted. This means that these versions exist within my local repository, but are not yet pushed to the main branch on GitHub. Let's say I've done a significant amount of work and I want to go ahead and commit and push to the main branch, here's what we'll do:

Open up GitHub Desktop, and on the left-hand side you'll see all of the files in your current local repo, which you can select or unselect if you don't want them to be pushed to GitHub (this is useful if you have large files > 25MB like data that you don't wanted pushed). You can also see the changes to each file you've added (in green) or removed (in red). Finally, on the bottom left-hand side you'll see where you are able to type in a commit message, along with a more detailed description explaining to others the changes you've made in this commit. Standard practice is to keep the commit message short and concise.



From here, just press the `Commit to main` and it will your edits will be committed; essentially this means that your edits are in a staging area, and that you've created a sort of checkpoint. You have to commit you edits before you can push. After commiting, you should see on the next page where you can `Push origin` in blue. When you're ready and sure that all of the edits you've committed are solid, go ahead and click the button to push the modified repo to GitHub.



After this we should see the repo being uploaded, and if we check out the GitHub repository online we'll see that those edits were pushed!



Now, this is a very basic tutorial that excludes all of the other etiquette that you need to learn when working on a project with other people in a shared repository. For example, if a team is working on the same code, typically each individual will create a seperate branch to work on, make edits, and make a pull request to merge with the main branch after it's been approved. 

So, there's much more to learn for sure, but this should get you started!
