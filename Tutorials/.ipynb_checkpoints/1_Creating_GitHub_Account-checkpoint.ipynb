### Welcome!

In this notebook we're going to focus on creating a GitHub account for work purposes. If you already have account for work purposes, then you can skip this tutorial entirely. It's important to note that non-NOAA GitHub accounts are not allowed to make contributions to NOAA Github repositories; if you're used to using your personal GitHub account, you'll need to transition.

Go ahead and head over to [GitHub.com](www.github.com), and click on the `sign up` button.



Go through the prompts by filling in your _*work email address*_, a username to be displayed to others, and your password. After this GitHub should be sending you a confirmation email to your work email address. Click on the link to verify and you should be presented with your repository!
